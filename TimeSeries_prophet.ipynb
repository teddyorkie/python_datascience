### Chuyên đề làm việc với kiểu datetime và chuyển đổi to/from string (o) type

Nếu ta check thử dtype của gm với lệnh `gm.index.dtype` ta có `dtype('<M8[ns]')`
Vậy thì: What is the difference between two types `datetime64[ns]` and `<M8[ns]` ?
Answer: `datetime64[ns]` is a general dtype, while `<M8[ns]` is a specific dtype. General dtypes map to specific dtypes, but may be different from one installation of NumPy to the next. On a machine whose byte order is little endian, there is no difference between np.dtype('datetime64[ns]') and np.dtype('<M8[ns]')

However, on a big endian machine, np.dtype('datetime64[ns]') would equal np.dtype('>M8[ns]').

So `datetime64[ns]` maps to either `<M8[ns]` or `>M8[ns]` depending on the endian-ness of the machine.
There are many other similar examples of general dtypes mapping to specific dtypes: `int64` maps to `<i8` or `>i8`, and `int` maps to either `int32` or `int64` depending on the bit architecture of the OS and how NumPy was compiled.

## Tính giá trị thị trường

```python
def add_cap_value(data_frame, yearly_shares_map):
    ###
    # data_frame must be having nsdatetime as index
    # we will add a new column 'cap' to the end and populate data
    # with formula shares * adj.close
    ###
    # Create a year column 
    data_frame['Year'] = data_frame.index.year

    # Take Dates from index and move to Date column 
    data_frame.reset_index(level=0, inplace = True)
    data_frame['cap'] = 0

    # Calculate market cap for all years
    for i, year in enumerate(data_frame['Year']):
        # Retrieve the shares for the year
        shares = yearly_shares_map.get(year)

        # Update the cap column to shares times the price
        data_frame.loc[i, 'cap'] = shares * data_frame.loc[i, 'Adj. Close']
        
    return data_frame

def merge_filter(gm_data, tesla_data):
    # Merge the two datasets and rename the columns
    cars = gm_data.merge(tesla_data, how='inner', on='Date')
    cars.rename(columns={'cap_x': 'gm_cap', 'cap_y': 'tesla_cap'}, inplace=True)
    
    # Select only the relevant columns
    cars = cars.loc[:, ['Date', 'gm_cap', 'tesla_cap']]

    # Divide to get market cap in billions of dollars
    cars['gm_cap'] = cars['gm_cap'] / 1e9
    cars['tesla_cap'] = cars['tesla_cap'] / 1e9
    return cars
```

```python
# use the above function to modify dataframe having new 'cap' column
tesla = add_cap_value(tesla, tesla_shares)
gm = add_cap_value(gm, gm_shares)
```

Merging là một phần thiết yếu trong khoa học dữ liệu vì nó cho phép ta kết hợp các bộ dữ liệu trên những cột chung. Trong trường hợp này, ta có giá cổ phiếu của hai công ty khác nhau vào cùng một ngày và do đó ta muốn join dữ liệu trên cột ngày. Chúng ta thực hiện merge "INNER" để chỉ lưu các mục mà ngày đó có dữ liệu trong cả hai phía. Sau khi merge, ta đổi tên các cột để biết cái nào đi với công ty xe hơi nào.

```python
cars = merge_filter(gm, tesla)
cars.head()
```

# Prophet Models 

[Thư viện Prophet do Facebook](https://facebook.github.io/prophet/docs/quick_start.html) phát triển được ra mắt vào 2017 cho Python và R. Prophet được thiết kế để phân tích chuỗi thời gian với các quan sát hàng ngày và hiển thị các mẫu trên các thang thời gian khác nhau. Nó cũng có các tính năng nâng cao để mô hình hóa các hiệu ứng của ngày lễ lên chuỗi thời gian và thực hiện các thay đổi tùy chỉnh, nhưng chúng ta sẽ bám sát các chức năng cơ bản ở đây. Prophet, giống như quandl, có thể được cài đặt với pip từ dòng lệnh.

Trước tiên, ta đổi tên các cột trong dữ liệu sang định dạng chính xác. Cột ngày phải được gọi là `ds` và cột giá trị mà ta muốn dự đoán `y`. Sau đó, ta tạo các mô hình và khớp chúng với dữ liệu, giống như mô hình học máy Scikit-Learn: 

```python
import fbprophet

# Prophet requires columns ds (Date) and y (value)
gm = gm.rename(columns={'Date': 'ds', 'cap': 'y'})
# Rescale market cap in billions
gm['y'] = gm['y'] / 1e9

# Make the prophet models and fit on the data
# changepoint_prior_scale can be changed to achieve a better fit
gm_prophet = fbprophet.Prophet(changepoint_prior_scale=0.05, daily_seasonality=True)
gm_prophet.fit(gm)

# Repeat for the tesla data
tesla =tesla.rename(columns={'Date': 'ds', 'cap': 'y'})
tesla['y'] = tesla['y'] / 1e9
tesla_prophet = fbprophet.Prophet(changepoint_prior_scale=0.05, n_changepoints=10, daily_seasonality=True)
tesla_prophet.fit(tesla);
```

Khi tạo một mô hình tiên tri, chúng ta cũng có thể chỉ định các điểm thay đổi, xảy ra khi một chuỗi thời gian chuyển từ tăng sang giảm hoặc từ tăng chậm sang tăng nhanh (nằm ở nơi thay đổi tốc độ trong chuỗi thời gian là lớn nhất). Thay đổi có thể tương ứng với các sự kiện quan trọng như ra mắt sản phẩm hoặc dao động kinh tế vĩ mô trên thị trường. Nếu chúng ta không chỉ định các điểm thay đổi, prophet sẽ tính chúng cho chúng ta.

## Predictions for 2 Years
Để thực hiện dự báo, chúng ta cần tạo chỗ cho dataframe tương lai như sau:

```python
# Make a future dataframe for 2 years
gm_forecast = gm_prophet.make_future_dataframe(periods=365 * 2, freq='D')
# Make predictions
gm_forecast = gm_prophet.predict(gm_forecast)

tesla_forecast = tesla_prophet.make_future_dataframe(periods=365*2, freq='D')
tesla_forecast = tesla_prophet.predict(tesla_forecast)
```

Vùng không chắc chắn nở rộng thêm trong tương lai vì sự không chắc chắn ban đầu lan truyền và phát triển theo thời gian. Điều này được thấy giống như trong các dự báo thời tiết càng trở nên kém chính xác hơn khi càng ra xa thời gian chúng được thực hiện.

```python
len(tesla_prophet.changepoints)
tesla_prophet.changepoints[:10]
```

Trên đây là danh sách 10 điểm chuyển đổi (changepoints) được mô hình tìm thấy. Điểm chuyển đổi thể hiện mốc thời gian mà chuỗi giá trị thay đổi đáng kể (Vd từ tăng sang giảm). Siêu tham số (hyperparameter) này được sử dụng để kiểm soát mức độ nhạy cảm của xu hướng đối với các thay đổi: các giá trị cao hơn là nhạy cảm hơn và giá trị thấp hơn ít nhạy cảm hơn. Giá trị này được sử dụng để chống lại một trong những sự đánh đổi cơ bản nhất trong học máy: [thiên vị so với phương sai](https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff).

Nếu ta học quá sát với dữ liệu đào tạo, sẽ bị gọi là overfitting (mô hình quá khớp), ta sẽ có quá nhiều dao động và mô hình của ta sẽ không thể khái quát tốt cho dữ liệu mới. Mặt khác, nếu mô hình của ta không nắm bắt được xu hướng trong dữ liệu đào tạo thì nó đang bị underfitting (chưa khớp) và sẽ có quá nhiều sai lệch. Khi một mô hình chưa khớp, việc tăng giá trị điểm chuyển đổi trước (changepoint prior) cho phép mô hình linh hoạt hơn để phù hợp với dữ liệu và nếu mô hình bị quá khớp, việc giảm giá trị này sẽ giới hạn tính linh hoạt. Hiệu ứng của thang đo thay đổi trước có thể được minh họa bằng các đồ thị dự đoán được thực hiện với một loạt các giá trị ở phần sau.

## Effect of Changepoint Prior Scale

```python
# Try 4 different changepoints
for changepoint in [0.001, 0.05, 0.1, 0.5]:
    model = fbprophet.Prophet(daily_seasonality=False, changepoint_prior_scale=changepoint)
    model.fit(tesla)
    
    future = model.make_future_dataframe(periods=365, freq='D')
    future = model.predict(future)
    
    tesla[changepoint] = future['yhat']
```

```python
# Create the plot
plt.figure(figsize=(10, 8))

# Actual observations
plt.plot(tesla['ds'], tesla['y'], 'ko', label = 'Observations')
colors = {0.001: 'b', 0.05: 'r', 0.1: 'grey', 0.5: 'gold'}

# Plot each of the changepoint predictions
for changepoint in [0.001, 0.05, 0.1, 0.5]:
    plt.plot(tesla['ds'], tesla[changepoint], color = colors[changepoint], label = '%.3f prior scale' % changepoint)
    
plt.legend(prop={'size': 14})
plt.xlabel('Date'); plt.ylabel('Market Cap (billions $)'); plt.title('Effect of Changepoint Prior Scale');
```
