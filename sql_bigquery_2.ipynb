### Ví dụ: Ngày nào trong tuần có nhiều tai nạn xe máy chết người nhất?
Lần này ta dùng database US Traffic Fatality Records, chứa dữ liệu về tai nạn giao thông tại Mỹ mà có ít nhất 1 người chết. Dùng bảng `accident_2015`, code sau xem 5 dòng đầu tiên của bảng.

````python
from google.cloud import bigquery

# Create a "Client" object
client = bigquery.Client()

# Construct a reference to the "nhtsa_traffic_fatalities" dataset
dataset_ref = client.dataset("nhtsa_traffic_fatalities", project="bigquery-public-data")
dataset = client.get_dataset(dataset_ref)

# Construct a reference to the "accident_2015" table
table_ref = dataset_ref.table("accident_2015")
table = client.get_table(table_ref)

# Preview the first five lines of the "accident_2015" table
client.list_rows(table, max_results=5).to_dataframe()
````

Xem bảng ta thấy:
* cột `consecutive_number` chứa ID duy nhất cho mỗi tai nạn và
* cột `timestamp_of_crash` chứa ngày của tai nạn theo định dạng DATETIME

vì vậy ta có thể
* **EXTRACT** ngày trong tuần từ cột `timestamp_of_crash` và
* **GROUP BY** ngày trong tuần trước khi **COUNT** cột `consecutive_number` để xác định số tai nạn cho mỗi ngày trong tuần.
* sau đó sắp xếp với **ORDER BY** theo số vụ tai nạn giảm dần

````python
# Query to find out the number of accidents for each day of the week
query = """
        SELECT COUNT(consecutive_number) AS num_accidents, 
               EXTRACT(DAYOFWEEK FROM timestamp_of_crash) AS day_of_week
        FROM `bigquery-public-data.nhtsa_traffic_fatalities.accident_2015`
        GROUP BY day_of_week
        ORDER BY num_accidents DESC
        """
````

````python
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=10**8)
query_job = client.query(query, job_config=safe_config)
accidents_by_day = query_job.to_dataframe()

# Print the DataFrame
accidents_by_day
````

Để biết ngày trả về này là thứ mấy, ta tham khảo [BigQuery document] (https://cloud.google.com/bigquery/docs/reference/legacy-sql#dayofweek) về hàm DAYOFWEEK.

## As & With
Với chỉ những gì chúng ta đã học, SQL query cũng đã có thể khá dài dòng, khiến cho việc đọc hiểu (để sửa lỗi) càng phức tạp. Ta có thể dùng **AS** và **WITH** để khiến câu lệnh sạch hơn, gọn hơn, dễ đọc hơn
### AS
AS còn được gọi là **aliasing** tương tự như `as` trong Python khi ta dùng `import pandas as pd`. Cách dùng trong SQL là ngay sau cột được select như đã thấy ở trên.
### WITH ... AS
**AS** kết hợp với **WITH** càng bá đạo hơn khi được dùng để tạo **CTE** hay **common table expression** tức là một bảng tạm được tạo ra trong query của bạn. CTEs hữu ích vì giúp chia query thành các khúc nhỏ dễ đọc hơn và ta có thể query tiếp trên chính chúng. Vd: với bảng sau
![table](https://i.imgur.com/MXrsiAZ.png)
ta có thể tạo bảng tạm với các con vật lớn tuổi rồi query trên bảng CTE đó
![CTE](https://i.imgur.com/3xQZM4p.png)

Ta có thể viết query này mà không dùng CTE nhưng nếu đây chỉ là một phần trong một query rất dài, bỏ qua CTE sẽ khiến câu lệnh rất khó đọc. *Lưu ý là CTEs chỉ tồn tại trong 1 query khác khi được tạo ra, và không thể được tham chiếu đến từ các query sau đó nữa.*
